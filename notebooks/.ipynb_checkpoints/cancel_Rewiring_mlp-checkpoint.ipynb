```julia
using Revise
includet("../codebase/mlp.jl")

using Knet, Plots, Statistics, LinearAlgebra, Random, .mlp
```

```julia
n=12 # length of input bit array
xdim = n
h1=h2=h3=2*n # hidden layer sizes
ydim = 1 # output is the probability for even parity
nbatch = 100 # number of bit strings presented at each iteration
ntest = div(2^n,8) # keep 1/8-th as test set
niter = 10^5 # number of training iterations
```

```julia
# NN parameters
w1 = Param(randn(h1,xdim))
b1 = Param(randn(h1))
w2 = Param(randn(h2,h1))
b2 = Param(randn(h2))
w3 = Param(randn(h3,h2))
b3 = Param(randn(h3))
w4 = Param(randn(1,h3))
b4 = Param(randn(1))
w = (w1,w2,w3,w4)
b = (b1,b2,b3,b4)
```

```julia
xtrn, ytrn, xtst, ytst = mlp.gen_and_split_data(n, ntest)
```

```julia
function backprop2(w, b, niter, n, ntest, nbatch; λ = 0.0, LR = 0.1, layer_dims = [], pRandPrune = 0.0)
    losstrn = []
    losstst = []
    xtrn, ytrn, xtst, ytst = mlp.gen_and_split_data(n, ntest)
    
    for k in progress(1:niter)
        xin = genbatch(n,nbatch, ntest)
        yin = sum(xin,dims=1).%2
        dl = @diff loss(w,b,xin,yin; λ= 0.001)
        for i=1:length(w)
            w[i] .-= LR *grad(dl,w[i])
            b[i] .-= LR *grad(dl,b[i])
        end
        
        if (k%2==1)
            flagAlg = "advection"
        else 
            flagAlg = "consensus"
        end
        
        A = computeKernel(w,layer_dims)
            
        ######### PRUNE ###########
        A_pruned, xCutHead, nodeX = prune(A, pRandPrune, 0)
        ######### REWIRE ##########
        # rewires tail nodes
        println(A_pruned)
        A_rewired = runInDynamics(A_pruned, pRandPrune, LR, flagAlg)
        println(sum(A_pruned.-A_rewired))
        w = computeKernel_reverse(w, A_rewired, layer_dims)
        
        
        if (k%1000==1)
            xin = genbatch(n,ntest, ntest) # training set samples with size equal to test set
            yin = sum(xin,dims=1).%2 
            push!(losstrn,loss(w,b,xin,yin; λ= 0.001)) # record loss over 1000 samples
            push!(losstst,loss(w,b,xtst,ytst; λ= 0.001))
        end
    end
    
    return w, b, losstrn, losstst
end
```

```julia
w, b, losstrn, losstst = backprop2(w, b, niter, n, ntest, nbatch; layer_dims = [xdim, h1, h2, h3, ydim], pRandPrune = 0.3);
```

```julia
# accuracy
function acc(w,b,x,y)
    cnt = 0
    tot = length(x[1,:])
    for i=1:tot-1
        chk = (pred(w,b,x[:,i])[1] - y[i][1])^2
        if (chk < 0.5)
            cnt += 1
        end
    end
    return 1.0*cnt/tot
end
```
